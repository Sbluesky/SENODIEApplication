```
data.describe()
```

```
df_aa = pd.DataFrame(columns=[0])
df_aa[0] = data[0].unique()
df_aa.describe()
```

```
data_test = df_aa
```

```
data_test = data_test[0].str.split('","', expand=True)
data_test.head(10)
```

```
#Create new dataframe
df_new = pd.DataFrame(columns=['time', 'user_id', 'product_id', 'belong_cate_lvl1_id', 'belong_cate_lvl2_id','belong_cate_lvl3_id','belong_cate_lvl1_name', 'belong_cate_lvl2_name','belong_cate_lvl3_name', 'href'])
```

```
#Chia cột
df_new['time'] = data_test[0].str.replace('{"time":"','')
df_new['user_id'] = data_test[1].str.replace('user_id":"','')
df_new['product_id'] = data_test[2].str.replace('product_id":"','')
df_new['belong_cate_lvl1_id'] = data_test[3].str.replace('belong_cate_lvl1_id":"','')
df_new['belong_cate_lvl2_id'] = data_test[4].str.replace('belong_cate_lvl2_id":"','')
df_new['belong_cate_lvl3_id'] = data_test[5].str.replace('belong_cate_lvl3_id":"','')
df_new['belong_cate_lvl1_name'] = data_test[6].str.replace('belong_cate_lvl1_name":"','')
df_new['belong_cate_lvl2_name'] = data_test[7].str.replace('belong_cate_lvl2_name":"','')
df_new['belong_cate_lvl3_name'] = data_test[8].str.replace('belong_cate_lvl3_name":"','')
df_new['href'] = data_test[9].str.replace('href":"','')
df_new.head()
```

```
#ép kiểu data
df_new['product_id'] = df_new['product_id'].astype('int64')
df_new['belong_cate_lvl1_id'] = df_new['belong_cate_lvl1_id'].astype('int32')
df_new['belong_cate_lvl2_id'] = df_new['belong_cate_lvl2_id'].astype('int32')
df_new['belong_cate_lvl3_id'] = df_new['belong_cate_lvl3_id'].astype('int32')
df_new['time'] = df_new['time'].astype('datetime64')
```
